```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
data=pd.read_csv('iris PCA.csv')
s=StandardScaler()
X=s.fit_transform(data)
X_corr=(1/150)*X.T.dot(X)
u,s,v=np.linalg.svd(X_corr)
eigvals,eigvects=s,u
variance=(eigvals/np.sum(eigvals))*100
plt.bar(range(5),variance,alpha=0.6)
plt.ylabel("% of explained vaiance")
plt.xlabel("Dimensions")
pc1 = X.dot(eigvects[:,0])
pc2 = X.dot(eigvects[:,1])
species =data.species
species=species.tolist()
def plot_scatter(pc1, pc2):
    fig, ax = plt.subplots(figsize=(15, 8))
    species_unique = list(set(species))
    species_colors = ["r","b","g"]
    for i, spec in enumerate(species):
        plt.scatter(pc1[i], pc2[i], label = spec, c=species_colors[species_unique.index(spec)])
        ax.annotate(str(i+1), (pc1[i],pc2[i]))
        ax.set_xlabel('Principal Component 1')
        ax.set_ylabel('Principal Component 2')
        plt.show()
plot_scatter(pc1, pc2)
```
